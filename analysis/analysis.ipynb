```python
import matplotlib.pyplot as plt
import json
```

# Selection method analysis

```python
character_class = "archer"

with open(character_class + "/selection/elite", "r") as fp:
    data = json.load(fp)
    elite_averages = data[0]
    elite_maxes = data[1]
    elite_diversities = data[2]

with open(character_class + "/selection/roulette", "r") as fp:
    data = json.load(fp)
    roulette_averages = data[0]
    roulette_maxes = data[1]
    roulette_diversities = data[2]

with open(character_class + "/selection/universal", "r") as fp:
    data = json.load(fp)
    universal_averages = data[0]
    universal_maxes = data[1]
    universal_diversities = data[2]

with open(character_class + "/selection/boltzmann", "r") as fp:
    data = json.load(fp)
    boltzmann_averages = data[0]
    boltzmann_maxes = data[1]
    boltzmann_diversities = data[2]

with open(character_class + "/selection/deterministic_tournament", "r") as fp:
    data = json.load(fp)
    deterministic_tournament_averages = data[0]
    deterministic_tournament_maxes = data[1]
    deterministic_tournament_diversities = data[2]

with open(character_class + "/selection/probabilistic_tournament", "r") as fp:
    data = json.load(fp)
    probabilistic_tournament_averages = data[0]
    probabilistic_tournament_maxes = data[1]
    probabilistic_tournament_diversities = data[2]

with open(character_class + "/selection/ranking", "r") as fp:
    data = json.load(fp)
    ranking_averages = data[0]
    ranking_maxes = data[1]
    ranking_diversities = data[2]
```

```python
plt.figure(figsize=(10,6))

plt.plot(elite_averages, label='elite')
plt.plot(roulette_averages, label='roulette')
plt.plot(universal_averages, label='universal')
plt.plot(boltzmann_averages, label='boltzmann')
plt.plot(deterministic_tournament_averages, label='deterministic_tournament')
plt.plot(probabilistic_tournament_averages, label='probabilistic_tournament')
#plt.plot(ranking_averages, label='ranking')

plt.xlabel('Generation')
plt.ylabel('Average fitness')

plt.title('Evolution of average fitness according to selection method')
plt.legend(title='Selection method')
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(elite_maxes, label='elite')
plt.plot(roulette_maxes, label='roulette')
plt.plot(universal_maxes, label='universal')
plt.plot(boltzmann_maxes, label='boltzmann')
plt.plot(deterministic_tournament_maxes, label='deterministic_tournament')
plt.plot(probabilistic_tournament_maxes, label='probabilistic_tournament')
#plt.plot(ranking_maxes, label='ranking')

plt.xlabel('Generation')
plt.ylabel('Max fitness encountered')

plt.title('Evolution of max fitness according to selection method')
plt.legend(title='Selection method')
plt.grid()

plt.show()
```

```python
plt.plot(elite_maxes)
plt.plot(elite_averages)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Evolution of average and max fitness in Elite selection')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(elite_diversities, label='elite')
plt.plot(roulette_diversities, label='roulette')
plt.plot(universal_diversities, label='universal')
plt.plot(universal_diversities, label='boltzmann')
plt.plot(deterministic_tournament_diversities, label='deterministic_tournament')
plt.plot(probabilistic_tournament_diversities, label='probabilistic_tournament')
#plt.plot(ranking_diversities, label='ranking')

plt.xlabel('Generation')
plt.ylabel('Diversity')

plt.title('Evolution of diversity of gene pool according to selection method')
plt.legend(title='Selection method')
plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

selection_methods = ['elite', 'roulette', 'universal', 'boltzmann', 'deterministic_tournament', 'probabilistic_tournament', 'ranking']

for selection_method in selection_methods:
    data_file = f"data_{selection_method}.csv"

    data_df = pd.read_csv(data_file, index_col='Generation')

    plt.figure(figsize=(10, 6))

    # Plot performances
    plt.subplot(2, 1, 1)
    plt.title(f"Average Performance and Diversity by Generation (Grouped by p_m) - {selection_method}")
    plt.xlabel("Generation")
    plt.ylabel("Average Performance")

    for column in data_df.columns:
        if column.startswith('Performance'):
            plt.plot(data_df.index[:150], data_df[column][:150], label=column)

    plt.legend()
    plt.grid()

    # Plot diversities
    plt.subplot(2, 1, 2)
    plt.xlabel("Generation")
    plt.ylabel("Diversity")

    for column in data_df.columns:
        if column.startswith('Diversity'):
            plt.plot(data_df.index[:150], data_df[column][:150], label=column)

    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```
